```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import cv2
```

```python
dt=pd.read_csv("train.csv")
dt.head()
```

```python
df=dt[["Smoke_image","x","y","w","h"]]
df.head()
```

```python
df["Smoke_image"]=df["Smoke_image"].apply(lambda x: x.split("/")[-1])
df.head()
```

```python
def coco_to_yolo(x,y,w,h,image_w=224,image_h=224):
    return f"15 {(2*x+w)/(2*image_w)} {(2*x+h)/(image_h)} {w/image_w} {h/image_h}"
```

```python
img_info={"image_w":[],"image_h":[]}
for i in df["Smoke_image"]:
    img=plt.imread(f"images/{i}")
    shape=img.shape
    img_info["image_h"].append(shape[1])
    img_info['image_w'].append(shape[0])
```

```python
df['file']=df.apply(lambda x: coco_to_yolo(x['x'],x['y'],x['w'],x['h']),axis=1)
```

```python
df.head()
```

```python
for i in range(len(df['Smoke_image'])):
    name=df['Smoke_image'][i][:7]
    data=df['file'][i]
    with open(f'labels/{name}.txt',"w") as file:
        file.writelines(data)
        file.close()
    if (i+1)%1000==0:
        print(i+1)
```
